# Decision Tree Loan Prediction Using Machine Learning
1. Objective: Predict whether a loan application will be approved based on customer data.

2. Dataset: Use a dataset containing features like income, credit history, employment status, loan amount, and property type.

3. Preprocessing: Clean the data, handle missing values, and encode categorical variables into numerical formats.

4. Algorithm: Apply a Decision Tree algorithm to classify loan approval outcomes.

5. Data Splitting: Divide the dataset into training and testing sets to evaluate model performance.

6. Model Training: Train the Decision Tree using the training data to find patterns in customer features.

7. Evaluation Metrics: Measure accuracy, precision, recall, and F1 score to assess model performance.

8. Visualization: Use tools like matplotlib to visualize the Decision Tree and understand decision-making paths.

9. Feature Importance: Identify which customer features most influence loan approval decisions.

10. Result Interpretation: Provide insights into why loans are approved or denied, helping financial institutions make better decisions.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data.drop('LoanID',axis=1,inplace=True)
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cat
```

```python
le = LabelEncoder()
```

```python
for i in cat:
    data[i] = le.fit_transform(data[i])
```

```python
data
```

```python
data.dtypes
```

```python
data.isna().sum()
```

```python
X_c = data.drop('Default',axis=1)
y_c = data[['Default']]
```

```python
y_c
```

```python
X_r = data.drop('LoanAmount',axis=1)
y_r = data[['LoanAmount']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
c_X_train,c_X_test,c_y_train,c_y_test = train_test_split(X_c,y_c,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC = DecisionTreeClassifier()
```

```python
DTC.fit(c_X_train,c_y_train)
```

```python
c_y_pred = DTC.predict(c_X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(c_y_test,c_y_pred)
```

```python
r_X_train,r_X_test,r_y_train,r_y_test = train_test_split(X_r,y_r,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
DTR = DecisionTreeRegressor()
```

```python
DTR.fit(r_X_train,r_y_train)
```

```python
r_y_pred = DTR.predict(r_X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(r_y_test,r_y_pred)
```

```python
a = pd.DataFrame(r_y_pred)
```
